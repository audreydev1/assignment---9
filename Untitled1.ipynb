# 1.5 Identifying outliers in continuous features and resolving them using IQR

```python
#we'll use a box plot to identify the outliers
plt.figure(figsize=(10,10))
sns.set_context("paper",font_scale=1.5)
sns.boxplot(data=heart[numerical_continuous])
plt.show()
```

```python
def treat_outliers(column):
    Q1=column.quantile(0.25)
    Q3=column.quantile(0.75)
    IQR=Q3-Q1
    threshold=IQR*1.5
    lower_limit=Q1-threshold
    upper_limit=Q3+threshold
    return lower_limit,upper_limit
```

```python
#calling the fxn to remove the outliers
for i in numerical_continuous:
    lower_limit,upper_limit=treat_outliers(heart[i])
    heart=heart.loc[(heart[i]>lower_limit) & (heart[i]<upper_limit)]
```

# 1.9 Slicing Data and Scaling Features

```python
#we'll first select the no.of features we're going to use chi_2 test 
chi2_check = []
for i in X.columns:
    if chi2_contingency(pd.crosstab(heart['target'], heart[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
res = pd.DataFrame(data = [X.columns, chi2_check] ).T 
res.columns = ['Column', 'Hypothesis']
print(res)
```
